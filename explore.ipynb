```python
import pandas as pd
import sys
import os

DATA_PATH = "march-machine-learning-mania-2025/"

def load_df_dict(path=DATA_PATH):
    data_dict = {}
    
    for fname in os.listdir(path):
        data_dict[fname.split('.')[0]] = pd.read_csv(path + fname)
        
    return data_dict
```

```python
data = load_df_dict()
```
